```python
import anndata
import os
import requests

save_path = "../../data/example_sce.h5ad"
save_path = "data/example_sce.h5ad"
if not os.path.exists(save_path):
    response = requests.get("https://go.wisc.edu/69435h")
    with open(save_path, "wb") as f:
        f.write(response.content)

example_sce = anndata.read_h5ad(save_path)
example_sce
```

```python
from scdesigner.minimal import PositiveNMF

formula = "~ bs(pseudotime, degree=5)"
model = PositiveNMF(formula, maxIter=500, epochs=50)
model.fit(example_sce, lr=2)
model.predict()
model.sample()
```
