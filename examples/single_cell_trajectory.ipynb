```python
import scvelo as scv

adata = scv.datasets.pancreas()
scv.pp.filter_genes(adata, min_shared_counts=20)
scv.pp.normalize_per_cell(adata)
scv.pp.filter_genes_dispersion(adata, n_top_genes=2000)
```

```python
from scdesigner.estimate import simulator
from scdesigner.margins.negative_binomial import NegativeBinomial

marginal = NegativeBinomial("~ clusters_coarse")
marginal
```

```python
margins = [(["Sntg1", "Snhg6"], NegativeBinomial("~ clusters_coarse", device="cpu"))]
sim = simulator(adata, margins, max_iter=5)
```

```python
margins = [(["Sntg1", "Snhg6"], NegativeBinomial({"alpha": "~ clusters_coarse"}, device="cpu"))]
sim = simulator(adata, margins, max_iter=5)
sim
```

```python
margins = [(["Sntg1", "Snhg6"], NegativeBinomial({"alpha": "~ clusters_coarse", "mu": "~ clusters_coarse"}, device="cpu"))]
simulator(adata, margins, max_iter=5)
```

```python
sim = simulator(
    adata, 
    [
        (["Sntg1", "Snhg6"], NegativeBinomial("~ clusters_coarse", device="cpu")),
        (["Ncoa2", "Sbspon", "Erdr1"], NegativeBinomial("~ 1", device="cpu"))
    ],
    max_iter=5, lr=1e-3
)
```
