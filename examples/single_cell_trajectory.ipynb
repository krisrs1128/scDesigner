```python
import scvelo as scv

adata = scv.datasets.pancreas()
scv.pp.filter_genes(adata, min_shared_counts=20)
scv.pp.filter_genes_dispersion(adata, n_top_genes=2000)
```

```python
from scdesigner.simulator import simulator
from scdesigner.margins.negative_binomial import NegativeBinomial

marginal = NegativeBinomial("~ clusters_coarse")
marginal
```

```python
margins = [(["Lactb2", "Snhg6"], NegativeBinomial("~ clusters_coarse"))]
sim = simulator(adata, margins)
```

```python
margins = [(["Lactb2", "Snhg6"], NegativeBinomial({"alpha": "~ clusters_coarse"}))]
sim = simulator(adata, margins)
sim
```

```python
margins = [(["Lactb2", "Snhg6"], NegativeBinomial({"alpha": "~ clusters_coarse", "mu": "~ clusters_coarse"}))]
simulator(adata, margins)
```

```python
sim = simulator(
    adata, 
    [
        (["Lactb2", "Snhg6"], NegativeBinomial("~ clusters_coarse")),
        (["Pkhd1", "Mcm3", "Prim2"], NegativeBinomial("~ 1"))
    ],
    max_iter=100
)
```

```python
sim.parameters()
```

```python
sim.predict(adata.obs)
```
